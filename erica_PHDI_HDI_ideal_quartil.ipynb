```python
# 'dataset' tem os dados de entrada para este script
import pandas as pd
import numpy as np
from IPython.display import display
from decimal import Decimal, getcontext



file_path= r'C:\Users\Usuário\Desktop\Python\bancodedados.csv'


# Criação do dataframe

df = pd.read_csv(file_path, delimiter=',', encoding='utf-8')
```

```python
import pandas as pd

# Lista de colunas nas quais queremos aplicar o cálculo de quartil
colunas = ['LQ', 'GORANDRO', 'GOGINOI', 'GORDT', 'PCR', 'HOMO', 'GLIC', 'TyG', 
           'ADIPONECTINAngmL', 'LEP', 'RBP4µgmL', 'QuemerinangmL', 'SODUmL', 
           'MDAµML', 'FRAPµMmL']

# Função para calcular quartil apenas para colunas com valores únicos suficientes
def calcular_quartil_seguro(coluna):
    if coluna.nunique() > 1:  # Verificar se há mais de 1 valor único
        return pd.qcut(coluna, q=4, labels=False, duplicates='drop')
    else:
        return pd.Series([0]*len(coluna))  # Se há apenas um valor único, retorna 0 para todas as linhas

# Aplicar o cálculo de quartil para cada coluna da lista, tratando duplicatas e valores nulos
for col in colunas:
    # Preencher NaN com o valor mínimo da coluna para garantir que qcut funcione corretamente
    df[col] = df[col].fillna(df[col].min())
    
    # Calcular quartis de forma segura
    df[f'{col}_quartil'] = calcular_quartil_seguro(df[col])

# Exibir a quantidade de valores únicos em cada coluna e os quartis calculados
for col in colunas:
    print(f'{col} - Valores únicos: {df[col].nunique()}')

# Exibir o DataFrame com as novas colunas de quartis
display(df)



```

```python


dfdinamica = df.groupby('PHDI_ideal_quartil').agg(
    LQ_média=('LQ', 'mean'),
    LQ_desvio=('LQ', 'std'),
    GORANDRO_média=('GORANDRO', 'mean'),
    GORANDRO_desvio=('GORANDRO', 'std'),
    GOGINOI_média=('GOGINOI', 'mean'),
    GOGINOI_desvio=('GOGINOI', 'std'),
    GORDT_média=('GORDT', 'mean'),
    GORDT_desvio=('GORDT', 'std'),
    PCR_média=('PCR', 'mean'),
    PCR_desvio=('PCR', 'std'),
    HOMO_média=('HOMO', 'mean'),
    HOMO_desvio=('HOMO', 'std'),
    GLIC_média=('GLIC', 'mean'),
    GLIC_desvio=('GLIC', 'std'),
    TyG_média=('TyG', 'mean'),
    TyG_desvio=('TyG', 'std'),
    ADIPONECTINAngmL_média=('ADIPONECTINAngmL', 'mean'),
    ADIPONECTINAngmL_desvio=('ADIPONECTINAngmL', 'std'),
    LEP_média=('LEP', 'mean'),
    LEP_desvio=('LEP', 'std'),
    RBP4µgmL_média=('RBP4µgmL', 'mean'),
    RBP4µgmL_desvio=('RBP4µgmL', 'std'),
    QuemerinangmL_média=('QuemerinangmL', 'mean'),
    QuemerinangmL_desvio=('QuemerinangmL', 'std'),
    SODUmL_média=('SODUmL', 'mean'),
    SODUmL_desvio=('SODUmL', 'std'),
    MDAµML_média=('MDAµML', 'mean'),
    MDAµML_desvio=('MDAµML', 'std'),
    FRAPµMmL_média=('FRAPµMmL', 'mean'),
    FRAPµMmL_desvio=('FRAPµMmL', 'std')
).round(2)




display(dfdinamica)
```
