```python
# 'dataset' tem os dados de entrada para este script
import pandas as pd
import numpy as np
from IPython.display import display
from decimal import Decimal, getcontext


df = pd.DataFrame()


# Número de elementos que você deseja gerar
num_elements = 50


# Gerar números flutuantes aleatórios entre 0 e 10 para cada coluna
columns = [
    'LQ', 'GORANDRO', 'GOGINOI', 'GORDT', 'PCR', 'HOMO', 'GLIC', 'TyG',
    'ADIPONECTINAngmL', 'LEP', 'RBP4µgmL', 'QuemerinangmL', 'SODUmL', 'MDAµML', 'FRAPµMmL',
    'PHDIquartil', 'PHDI_aqd_quartil', 'PHDI_mod_quartil', 'PHDI_razao_quartil', 'HDI_ideal_quartil'
]


for column in columns:
    df[column] = np.random.uniform(0, 10, num_elements)

print(df)
```

```python
df['HDI_ideal_quartil'] = pd.qcut(df['LQ'], q=4, labels=False)
print(df.columns)

print(df)
```

```python


dfdinamica = df.groupby('HDI_ideal_quartil').agg(
    LQ_média=('LQ', 'mean'),
    LQ_desvio=('LQ', 'std'),
    GORANDRO_média=('GORANDRO', 'mean'),
    GORANDRO_desvio=('GORANDRO', 'std'),
    GOGINOI_média=('GOGINOI', 'mean'),
    GOGINOI_desvio=('GOGINOI', 'std'),
    GORDT_média=('GORDT', 'mean'),
    GORDT_desvio=('GORDT', 'std'),
    PCR_média=('PCR', 'mean'),
    PCR_desvio=('PCR', 'std'),
    HOMO_média=('HOMO', 'mean'),
    HOMO_desvio=('HOMO', 'std'),
    GLIC_média=('GLIC', 'mean'),
    GLIC_desvio=('GLIC', 'std'),
    TyG_média=('TyG', 'mean'),
    TyG_desvio=('TyG', 'std'),
    ADIPONECTINAngmL_média=('ADIPONECTINAngmL', 'mean'),
    ADIPONECTINAngmL_desvio=('ADIPONECTINAngmL', 'std'),
    LEP_média=('LEP', 'mean'),
    LEP_desvio=('LEP', 'std'),
    RBP4µgmL_média=('RBP4µgmL', 'mean'),
    RBP4µgmL_desvio=('RBP4µgmL', 'std'),
    QuemerinangmL_média=('QuemerinangmL', 'mean'),
    QuemerinangmL_desvio=('QuemerinangmL', 'std'),
    SODUmL_média=('SODUmL', 'mean'),
    SODUmL_desvio=('SODUmL', 'std'),
    MDAµML_média=('MDAµML', 'mean'),
    MDAµML_desvio=('MDAµML', 'std'),
    FRAPµMmL_média=('FRAPµMmL', 'mean'),
    FRAPµMmL_desvio=('FRAPµMmL', 'std')
).round(2)




display(dfdinamica)
```
