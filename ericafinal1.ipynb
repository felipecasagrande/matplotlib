```python
# 'dataset' tem os dados de entrada para este script
import pandas as pd
import numpy as np
from IPython.display import display
from decimal import Decimal, getcontext

PHDI_castanhas= np.random.uniform(0,10,100)

df=pd.DataFrame({'df':PHDI_castanhas})


# Número de elementos que você deseja gerar
num_elements = 100


# Gerar números flutuantes aleatórios entre 0 e 10 para cada coluna
columns = [
    'PHDI_castanhas','PHDI_leguminosas', 'PHDI_fruta', 'PHDI_vegetais', 'PHDI_cereais',
    'PHDI_ovo', 'PHDI_peixe', 'PHDI_tuber', 'PHDI_lacteos', 'PHDI_oleos',
    'PHDI_veg_verdes', 'PHDI_veg_verm', 'PHDI_carne', 'PHDI_frango',
    'PHDI_saturada', 'PHDI_acucar'
]

for column in columns:
    df[column] = np.random.uniform(0, 10, num_elements)
    df['dislipidemia']=np.random.randint(0,5,100)
display(df)
```

```python




dfdinamica = df.groupby('dislipidemia').agg(
    PHDI_castanhas_média=('PHDI_castanhas', 'mean'),
    PHDI_castanhas_desvio=('PHDI_castanhas', 'std'),
    PHDI_leguminosas_média=('PHDI_leguminosas', 'mean'),
    PHDI_leguminosas_desvio=('PHDI_leguminosas', 'std'),
    PHDI_fruta_média=('PHDI_fruta', 'mean'),
    PHDI_fruta_desvio=('PHDI_fruta', 'std'),
    PHDI_vegetais_média=('PHDI_vegetais', 'mean'),
    PHDI_vegetais_desvio=('PHDI_vegetais', 'std'),
    PHDI_cereais_média=('PHDI_cereais', 'mean'),
    PHDI_cereais_desvio=('PHDI_cereais', 'std'),
    PHDI_ovo_média=('PHDI_ovo', 'mean'),
    PHDI_ovo_desvio=('PHDI_ovo', 'std'),
    PHDI_peixe_média=('PHDI_peixe', 'mean'),
    PHDI_peixe_desvio=('PHDI_peixe', 'std'),
    PHDI_tuber_média=('PHDI_tuber', 'mean'),
    PHDI_tuber_desvio=('PHDI_tuber', 'std'),
    PHDI_lacteos_média=('PHDI_lacteos', 'mean'),
    PHDI_lacteos_desvio=('PHDI_lacteos', 'std'),
    PHDI_oleos_média=('PHDI_oleos', 'mean'),
    PHDI_oleos_desvio=('PHDI_oleos', 'std'),
    PHDI_veg_verdes_média=('PHDI_veg_verdes', 'mean'),
    PHDI_veg_verdes_desvio=('PHDI_veg_verdes', 'std'),
    PHDI_veg_verm_média=('PHDI_veg_verm', 'mean'),
    PHDI_veg_verm_desvio=('PHDI_veg_verm', 'std'),
    PHDI_carne_média=('PHDI_carne', 'mean'),
    PHDI_carne_desvio=('PHDI_carne', 'std'),
    PHDI_frango_média=('PHDI_frango', 'mean'),
    PHDI_frango_desvio=('PHDI_frango', 'std'),
    PHDI_saturada_média=('PHDI_saturada', 'mean'),
    PHDI_saturada_desvio=('PHDI_saturada', 'std'),
    PHDI_acucar_média=('PHDI_acucar', 'mean'),
    PHDI_acucar_desvio=('PHDI_acucar', 'std')
).round(2)



print(dfdinamica)
```

```python

# Supondo que dfdinamica já esteja definido e as colunas estejam calculadas corretamente
dfdinamica['PHDI_castanhas_final'] = (
    dfdinamica['PHDI_castanhas_média'].astype(str) +
    ' (' + dfdinamica['PHDI_castanhas_desvio'].astype(str) + ')'
)
# Concatenar médias e desvios padrão para todas as colunas
dfdinamica['PHDI_leguminosas_final'] = (
    dfdinamica['PHDI_leguminosas_média'].astype(str) +
    ' (' + dfdinamica['PHDI_leguminosas_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_fruta_final'] = (
    dfdinamica['PHDI_fruta_média'].astype(str) +
    ' (' + dfdinamica['PHDI_fruta_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_vegetais_final'] = (
    dfdinamica['PHDI_vegetais_média'].astype(str) +
    ' (' + dfdinamica['PHDI_vegetais_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_cereais_final'] = (
    dfdinamica['PHDI_cereais_média'].astype(str) +
    ' (' + dfdinamica['PHDI_cereais_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_ovo_final'] = (
    dfdinamica['PHDI_ovo_média'].astype(str) +
    ' (' + dfdinamica['PHDI_ovo_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_peixe_final'] = (
    dfdinamica['PHDI_peixe_média'].astype(str) +
    ' (' + dfdinamica['PHDI_peixe_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_tuber_final'] = (
    dfdinamica['PHDI_tuber_média'].astype(str) +
    ' (' + dfdinamica['PHDI_tuber_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_lacteos_final'] = (
    dfdinamica['PHDI_lacteos_média'].astype(str) +
    ' (' + dfdinamica['PHDI_lacteos_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_oleos_final'] = (
    dfdinamica['PHDI_oleos_média'].astype(str) +
    ' (' + dfdinamica['PHDI_oleos_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_veg_verdes_final'] = (
    dfdinamica['PHDI_veg_verdes_média'].astype(str) +
    ' (' + dfdinamica['PHDI_veg_verdes_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_veg_verm_final'] = (
    dfdinamica['PHDI_veg_verm_média'].astype(str) +
    ' (' + dfdinamica['PHDI_veg_verm_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_carne_final'] = (
    dfdinamica['PHDI_carne_média'].astype(str) +
    ' (' + dfdinamica['PHDI_carne_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_frango_final'] = (
    dfdinamica['PHDI_frango_média'].astype(str) +
    ' (' + dfdinamica['PHDI_frango_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_saturada_final'] = (
    dfdinamica['PHDI_saturada_média'].astype(str) +
    ' (' + dfdinamica['PHDI_saturada_desvio'].astype(str) + ')'
)

dfdinamica['PHDI_acucar_final'] = (
    dfdinamica['PHDI_acucar_média'].astype(str) +
    ' (' + dfdinamica['PHDI_acucar_desvio'].astype(str) + ')'
)

display(dfdinamica)
```
