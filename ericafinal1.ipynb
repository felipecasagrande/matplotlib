```python
# 'dataset' tem os dados de entrada para este script
import pandas as pd
import numpy as np
from IPython.display import display
from decimal import Decimal, getcontext

PHDI_castanhas= np.random.uniform(0,10,100)

df=pd.DataFrame({'df':PHDI_castanhas})


# NÃºmero de elementos que vocÃª deseja gerar
num_elements = 100


# Gerar nÃºmeros flutuantes aleatÃ³rios entre 0 e 10 para cada coluna
columns = [
    'PHDI_castanhas','PHDI_leguminosas', 'PHDI_fruta', 'PHDI_vegetais', 'PHDI_cereais',
    'PHDI_ovo', 'PHDI_peixe', 'PHDI_tuber', 'PHDI_lacteos', 'PHDI_oleos',
    'PHDI_veg_verdes', 'PHDI_veg_verm', 'PHDI_carne', 'PHDI_frango',
    'PHDI_saturada', 'PHDI_acucar'
]

for column in columns:
    df[column] = np.random.uniform(0, 10, num_elements)
    df['dislipidemia']=np.random.randint(0,5,100)
display(df)
```
